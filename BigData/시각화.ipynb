```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)


df = pd.read_csv("설문지.csv")
```

```python
display(df)
print(df.isnull().sum())
df.info()
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.countplot(x = '1. 패션에 어느정도 관심이 있나요?', data = df, ax = ax1)
sns.countplot(x = '1-1. 패션의 유행도에 민감한가요?', hue = '1. 패션에 어느정도 관심이 있나요?', data = df, ax=ax2)

plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
ax3 = fig.add_subplot(1,2,1)
ax4 = fig.add_subplot(1,2,2)

sns.countplot(x = '2. 아침에 옷을 고르는데 시간이 얼마나 걸리나요?', data = df, ax = ax3)
sns.countplot(x = '2-1. 10분 이상 걸린다면 왜그런가요?', hue = '2. 아침에 옷을 고르는데 시간이 얼마나 걸리나요?', data = df, ax=ax4)

plt.show()
```

```python
fig = plt.figure(figsize=(20,15))
ax5 = fig.add_subplot(2,2,1)
ax6 = fig.add_subplot(2,2,2)
ax7 = fig.add_subplot(2,2,3)
ax8 = fig.add_subplot(2,2,4)

sns.countplot(x = '3. 타인에게 자신의 패션에 대해 질문한 적이 있나요?', data = df, ax = ax5)
sns.countplot(x = '3-1. 타인에게 옷을 추천 받은 적이 있나요?', data = df, ax=ax6)
sns.countplot(x = '3-2. 추천을 받았다면 그 추천에 만족하시나요?', hue = '3-1. 타인에게 옷을 추천 받은 적이 있나요?', data = df, ax=ax7)
sns.countplot(x = '3-3. 왜 추천을 받지 않았나요?', hue = '3-1. 타인에게 옷을 추천 받은 적이 있나요?', data = df, ax=ax8)

plt.show()
```

```python
fig = plt.figure(figsize=(5,20))
ax8 = fig.add_subplot(3,1,1)
ax9 = fig.add_subplot(3,1,2)
ax10 = fig.add_subplot(3,1,3)

sns.countplot(x = '4. 옷방이나 옷장이 습하거나 너무 건조해서 옷이 오염된 경우가 있나요?', data = df, ax = ax8)
sns.countplot(x = '본인이 가지고있는 옷을 잊고 비슷한 옷을 구매한적이 있나요?', data = df, ax = ax9)
sns.countplot(x = '트렌드에 맞는 패션 정보를 제공해주는 스마트 옷장이 생긴다면 사용할 의사가 있나요?', data = df, ax = ax10)

plt.show()
```
